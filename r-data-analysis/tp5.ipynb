```r
#2
varx = c(0, 0.2, 0.3, 0.6)
vary = c(1.01, 1.44, 1.55, 2.1)
plot(varx,vary)
abline(a=1.033,b=1.789,col="red")
```

```r
#3
# re = sum(vary-(varx*1.789+1.033))
# mc = ((varx*1.789+1.033) - mean(vary))^2

m$residuals
(m$fitted.values - mean(vary))^2
```

```r
#4
S_y = sum((vary - mean(vary))^2)/length(varx)
S_reg = sum((m$fitted.values - mean(vary))^2)/length(varx)
S_res = sum((vary - m$fitted.values)^2)/length(varx)

a = S_y - (S_reg + S_res)
R_2 = S_reg/S_y
cor_pearson = cor.test(varx,vary)

cor_pearson
R_2
```
